```python
# Ensure date_time is datetime type
df1_df2['date_time'] = pd.to_datetime(df1_df2['date_time'])

# Sort by visit_id and date_time to get proper step order per visit
df1_df2 = df1_df2.sort_values(['visit_id', 'date_time'])

# Calculate next step's timestamp per visit
df1_df2['next_date_time'] = df1_df2.groupby('visit_id')['date_time'].shift(-1)

# Calculate time spent on current step (seconds)
df1_df2['time_spent_sec'] = (df1_df2['next_date_time'] - df1_df2['date_time']).dt.total_seconds()

# Filter out invalid or missing durations (e.g., last step per visit has no next_date_time)
valid_time_spent = df1_df2['time_spent_sec'].notna() & (df1_df2['time_spent_sec'] > 0)
time_spent_df = df1_df2.loc[valid_time_spent, ['Variation', 'process_step', 'time_spent_sec']]

# Calculate average time spent by Variation and step
avg_time_per_step = time_spent_df.groupby(['Variation', 'process_step']).agg(
    avg_time_spent_sec=('time_spent_sec', 'mean'),
    count_samples=('time_spent_sec', 'count')
).reset_index()

print(avg_time_per_step)
```

```python
# Make sure date_time is datetime type
df1_df2['date_time'] = pd.to_datetime(df1_df2['date_time'])

# Map steps to numeric order for comparison
step_order_map = {'start': 0, 'step_1': 1, 'step_2': 2, 'step_3': 3, 'confirm': 4}
df1_df2['step_order'] = df1_df2['process_step'].map(step_order_map)

# Sort by visit_id and date_time to keep step order
df1_df2 = df1_df2.sort_values(['visit_id', 'date_time'])

# Get the next step order within the same visit
df1_df2['next_step_order'] = df1_df2.groupby('visit_id')['step_order'].shift(-1)

# Flag backward steps where next step order is less than current step order
df1_df2['is_backward_step'] = (df1_df2['next_step_order'] < df1_df2['step_order']).astype(int)

# Aggregate to visit level: any backward step means error for that visit
visit_error = df1_df2.groupby(['visit_id', 'Variation'])['is_backward_step'].max().reset_index()

# Calculate error rate per Variation (fraction of visits with at least one backward step)
error_rate = visit_error.groupby('Variation')['is_backward_step'].mean().reset_index()
error_rate = error_rate.rename(columns={'is_backward_step': 'error_rate'})

print(error_rate)
```

```python
# Assume df1_df2 is loaded and date_time parsed
df1_df2['date_time'] = pd.to_datetime(df1_df2['date_time'])

# Map steps to numeric order
step_order_map = {'start': 0, 'step_1': 1, 'step_2': 2, 'step_3': 3, 'confirm': 4}
df1_df2['step_order'] = df1_df2['process_step'].map(step_order_map)

# --- Completion Rate ---
visit_max_step = df1_df2.groupby(['visit_id', 'Variation'])['step_order'].max().reset_index()
visit_max_step['completed'] = visit_max_step['step_order'] == 4
completion_rate = visit_max_step.groupby('Variation')['completed'].mean().reset_index()
completion_rate.rename(columns={'completed': 'completion_rate'}, inplace=True)

# --- Average Time Spent per Step ---
df1_df2 = df1_df2.sort_values(['visit_id', 'date_time'])
df1_df2['next_date_time'] = df1_df2.groupby('visit_id')['date_time'].shift(-1)
df1_df2['time_spent_sec'] = (df1_df2['next_date_time'] - df1_df2['date_time']).dt.total_seconds()
valid_time_spent = df1_df2['time_spent_sec'].notna() & (df1_df2['time_spent_sec'] > 0)
time_spent_df = df1_df2.loc[valid_time_spent, ['Variation', 'process_step', 'time_spent_sec']]
avg_time_per_step = time_spent_df.groupby(['Variation', 'process_step']).agg(
    avg_time_spent_sec=('time_spent_sec', 'mean'),
    count_samples=('time_spent_sec', 'count')
).reset_index()

# --- Error Rate ---
df1_df2 = df1_df2.sort_values(['visit_id', 'date_time'])
df1_df2['next_step_order'] = df1_df2.groupby('visit_id')['step_order'].shift(-1)
df1_df2['is_backward_step'] = (df1_df2['next_step_order'] < df1_df2['step_order']).astype(int)
visit_error = df1_df2.groupby(['visit_id', 'Variation'])['is_backward_step'].max().reset_index()
error_rate = visit_error.groupby('Variation')['is_backward_step'].mean().reset_index()
error_rate.rename(columns={'is_backward_step': 'error_rate'}, inplace=True)

# --- Save to Excel ---
with pd.ExcelWriter('ab_test_summary.xlsx') as writer:
    completion_rate.to_excel(writer, sheet_name='Completion Rate', index=False)
    avg_time_per_step.to_excel(writer, sheet_name='Avg Time per Step', index=False)
    error_rate.to_excel(writer, sheet_name='Error Rate', index=False)

print("Excel file 'ab_test_summary.xlsx' created with all KPIs.")
```

```python
# Assume df1_df2 is your DataFrame and date_time is already parsed
df1_df2['date_time'] = pd.to_datetime(df1_df2['date_time'])

# Map steps to numeric order
step_order_map = {'start': 0, 'step_1': 1, 'step_2': 2, 'step_3': 3, 'confirm': 4}
df1_df2['step_order'] = df1_df2['process_step'].map(step_order_map)

# --- Completion Rate ---
visit_max_step = df1_df2.groupby(['visit_id', 'Variation'])['step_order'].max().reset_index()
visit_max_step['completed'] = visit_max_step['step_order'] == 4
completion_rate = visit_max_step.groupby('Variation')['completed'].mean().reset_index()
completion_rate.rename(columns={'completed': 'completion_rate'}, inplace=True)
completion_rate['process_step'] = None  # Add process_step column for uniformity

# --- Average Time Spent per Step ---
df1_df2 = df1_df2.sort_values(['visit_id', 'date_time'])
df1_df2['next_date_time'] = df1_df2.groupby('visit_id')['date_time'].shift(-1)
df1_df2['time_spent_sec'] = (df1_df2['next_date_time'] - df1_df2['date_time']).dt.total_seconds()
valid_time_spent = df1_df2['time_spent_sec'].notna() & (df1_df2['time_spent_sec'] > 0)
time_spent_df = df1_df2.loc[valid_time_spent, ['Variation', 'process_step', 'time_spent_sec']]
avg_time_per_step = time_spent_df.groupby(['Variation', 'process_step']).agg(
    avg_time_spent_sec=('time_spent_sec', 'mean'),
    count_samples=('time_spent_sec', 'count')
).reset_index()

# --- Error Rate ---
df1_df2 = df1_df2.sort_values(['visit_id', 'date_time'])
df1_df2['next_step_order'] = df1_df2.groupby('visit_id')['step_order'].shift(-1)
df1_df2['is_backward_step'] = (df1_df2['next_step_order'] < df1_df2['step_order']).astype(int)
visit_error = df1_df2.groupby(['visit_id', 'Variation'])['is_backward_step'].max().reset_index()
error_rate = visit_error.groupby('Variation')['is_backward_step'].mean().reset_index()
error_rate.rename(columns={'is_backward_step': 'error_rate'}, inplace=True)
error_rate['process_step'] = None  # Add process_step column for uniformity

# --- Combine all KPI DataFrames ---
combined_df = pd.merge(completion_rate, error_rate, on=['Variation', 'process_step'], how='outer')
combined_df = pd.merge(combined_df, avg_time_per_step, on=['Variation', 'process_step'], how='outer')

# --- Save combined data to Excel ---
combined_df.to_excel('ab_test_all_kpis_combined.xlsx', index=False)

print("Combined KPI Excel 'ab_test_all_kpis_combined.xlsx' created successfully.")
```
