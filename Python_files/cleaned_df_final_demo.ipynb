```python
df_final_demo.isnull().sum() 
```

```python
# Remove missing / unknown values

df_final_demo['gendr'].replace('U', pd.NA)

# Check if there are still U values in 'gendr' column
df_final_demo['gendr'].value_counts()
# Replace U values with NaN
df_final_demo['gendr'].replace('U', pd.NA, inplace=True)
# Check if there are still U values in 'gendr' column
df_final_demo['gendr'].value_counts()

```

```python
# Replace X values with NaN
df_final_demo['gendr'].replace('X', pd.NA, inplace=True)
# Check if there are still X values in 'gendr' column
df_final_demo['gendr'].value_counts()
```

```python
#Check for duplicates

df_final_demo.duplicated().sum()
```

```python
#Check for unvalid numerica data

print(df_final_demo[df_final_demo['clnt_age']<0])
print(df_final_demo[df_final_demo['clnt_age']>120])
print(df_final_demo[df_final_demo['bal']<0])
```

```python
# unique values for each column
for column in df_final_demo.columns:
    print(f"{column}: {df_final_demo[column].nunique()} unique values")
```

```python
# count Nan values for each column 

nan_counts = df_final_demo.isna().sum()
print(nan_counts)
```

```python
#Drop rows with any NaN values 
df_final_demo.dropna(inplace=True)
# Check if there are still NaN values
print(df_final_demo.isna().sum())

```
